```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.arange(0, 10)
y1 = x1 ** 2

plt.plot(x1, y1)
plt.title("y1 = x1^2")
plt.xlabel("x1")
plt.ylabel("y1")
plt.show()
```

```python
x2 = np.random.rand(50)
y2 = np.random.rand(50)

plt.scatter(x2, y2)
plt.title("Random Scatter Plot")
plt.xlabel("x2")
plt.ylabel("y2")
plt.show()
```

```python
labels = ["A", "B", "C", "D", "E"]
values = [10, 7, 3, 8, 5]

plt.bar(labels, values)
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
plt.Figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Plot1")

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title("Plot2")

plt.show()
```
