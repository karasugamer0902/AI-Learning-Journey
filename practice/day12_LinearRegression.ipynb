```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv("cleaned_data.csv")
df
```

```python
X = df[["age"]]
y = df["成績"]

model = LinearRegression()
model.fit(X , y)
print("Finished")
```

```python
test_df = pd.DataFrame({"age": [30, 50]})
predictions = model.predict(test_df)

print(f"預測 30 歲成績：{predictions[0]:.2f}")
print(f"預測 50 歲成績：{predictions[1]:.2f}")
```

```python
plt.Figure(figsize=(8 , 6))

plt.scatter(df["age"], df["成績"], color = 'blue', label = 'Actual Data')

plt.plot(df["age"], model.predict(df[["age"]]), color = 'red', linewidth = 2, label = 'Regression Line')

plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Age v.s Score - Linear Regression")
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```

```python
print(f"年齡每增加一歲，分數會改變：{model.coef_[0]:.2f} 分")
print(f"基礎分數（截距）是：{model.intercept_:.2f} 分")
```

```python
import joblib

joblib.dump(model, 'linear_regression_v1.pkl')
print("模型 v1 已儲存成功！")
```
