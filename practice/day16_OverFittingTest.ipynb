```python
import pandas as pd
import joblib
from sklearn.linear_model import LinearRegression

model1 = joblib.load('linear_regression_v1.pkl')
model2 = joblib.load('linear_regression_v2.pkl')
df = pd.read_csv("cleaned_data.csv")
df_chaos = pd.read_csv("data_chaos.csv")
X_multi = df_chaos[["age" , "hours"]]
y_score = df_chaos["成績"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_multi, y_score, test_size=0.2, random_state=42)

print(f"訓練用資料量: {len(X_train)} 筆")
print(f"測試用資料量: {len(X_test)} 筆")
```

```python
model_split = LinearRegression()
model_split.fit(X_train, y_train)

train_r2 = model_split.score(X_train, y_train)
test_r2 = model_split.score(X_test, y_test)

print(f"訓練集 R2 分數: {train_r2:.4f}")
print(f"測試集 R2 分數: {test_r2:.4f}")
```
