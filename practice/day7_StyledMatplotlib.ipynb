```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("classic")
x = np.arange(0, 10)
y1 = x ** 2
y2 = x * 5

plt.grid(True)

plt.plot(x, y1, color="cyan", linewidth=2, marker="o", linestyle="-", label="x^2")
plt.plot(x, y2, color="red", linewidth=2, marker="s", linestyle="--", label="x*5")
plt.plot(x, np.sqrt(x), color="green", linewidth=2, marker="^", linestyle=":", label="np,sqrt(x)")

plt.legend()

plt.show()
```
