```python
import numpy as np
data1 = np.random.normal(50, 15, size=200)
data1[:10]
```

```python
mean = np.mean(data1)
median = np.median(data1)
std = np.std(data1)

mean, median, std
```

```python
import matplotlib.pyplot as plt
plt.hist(data1, bins=15, color="skyblue", edgecolor="black")
plt.title("Distribution of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data2 = np.random.normal(50, 15, size=200)

mean = np.mean(data2)
std = np.std(data2)

x = np.arange(len(data2))
y = np.sort(data2)

plt.figure(figsize=(10, 5))

plt.plot(x, y, label="Data", color="blue")
plt.fill_between(x, mean-std, mean+std, color="yellow", alpha=0.3, label="Mean ± Std")

plt.axhline(mean, color="red", linestyle="--", label="Mean")

plt.legend()
plt.title("Data with Mean ± Std")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
plt.boxplot(data2, vert=True, patch_artist=True,
            boxprops=dict(facecolor="skyblue", color="blue"),
            medianprops=dict(color="red"))
plt.title("Boxplot of Data")
plt.ylabel("Value")
plt.show()
```

```python
from scipy.stats import gaussian_kde

kde = gaussian_kde(data2)

xs = np.linspace(min(data2), max(data2), 200)
ys = kde(xs)

plt.plot(xs, ys, color="green", linewidth=3)
plt.title("Kde of Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
