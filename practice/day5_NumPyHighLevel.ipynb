```python
import numpy as np
```

```python
x = np.array([5, 12, 3, 20, 10, 6])
np.where(x > 10 , 999, -1)
```

```python
print(np.argmax(x))
print(np.argmin(x))
```

```python
print(np.sort(x))
print("===============")
print(np.argsort(x))
print("===============")
print(np.argsort(x)[-5:])
```

```python
y = np.array([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])

print(np.unique(y))
print(np.unique(y, return_counts = True))
```

```python
print(np.bincount(y))
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.stack([a, b]))
print("===============")
print(np.hstack([a, b]))
print("===============")
print(np.vstack([a, b]))
```
