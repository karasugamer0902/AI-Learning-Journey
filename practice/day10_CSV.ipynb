```python
import pandas as pd
import numpy as np
```

```python
raw = pd.DataFrame({
    "name": ["Eric", "gpt", "A", "B", None],
    "age": [24, 10, None, 66, 67],
    "score": [430, None, 56, 16.6, None],
    "city": ["1", "2", None, "4", "5"]
})
raw
```

```python
raw.isna().sum()
```

```python
raw["score"] = raw["score"].fillna(raw['score'].mean())
raw
```

```python
raw["city"] = raw["city"].fillna(0)
raw["city"] = raw["city"].astype(int)
print(raw.dtypes)
raw
```

```python
result = raw[(raw["age"] > 20) & (raw["score"] > raw["score"].mean())]
result
```

```python
raw.sort_values(by = "score" , ascending=False)
```

```python
raw["status"] = np.where(raw["score"] >= 60 , "Pass" , "Fail")
raw
```

```python
fill_values = {"name":"Anonymous" , "age":raw["age"].median()}
raw = raw.fillna(value = fill_values)
raw
```

```python
raw.groupby("status")["age"].mean()
```

```python
raw = raw.round({"age":0 , "score":1})
raw
```

```python
raw = raw.rename(columns = {"name":"姓名" , "score":"成績"})
raw
```

```python
raw.to_csv("cleaned_data.csv", index=False)
```
