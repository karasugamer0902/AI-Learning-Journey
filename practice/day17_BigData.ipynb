```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

np.random.seed(42)
ages = np.random.randint(10 , 81 , size = 100)
hours = np.random.uniform(0 , 20 , size = 100)
scores = (hours * 25) - (ages * 1) + np.random.normal(50, 20, size=100)

df_big = pd.DataFrame({
    'age': ages,
    'hours': hours,
    'score': scores
})

print("100筆數據已生成")
df_big.head()
```

```python
X_big = df_big[['age' , 'hours']]
y_big = df_big['score']

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_big, y_big, test_size=0.2, random_state=42)

model_big = LinearRegression()
model_big.fit(X_train_b, y_train_b)

print(f"百人班級 - 訓練集 R2: {model_big.score(X_train_b, y_train_b):.4f}")
print(f"百人班級 - 測試集 R2: {model_big.score(X_test_b, y_test_b):.4f}")
```

```python
import joblib

joblib.dump(model_big, '100people_score_model.pkl')
```
