```python
import pandas as pd
import joblib
from sklearn.metrics import r2_score
from sklearn.linear_model import LogisticRegression

model1 = joblib.load('linear_regression_v1.pkl')
model2 = joblib.load('linear_regression_v2.pkl')
df = pd.read_csv("cleaned_data.csv")
df_chaos = pd.read_csv("data_chaos.csv")
```

```python
r2_v1 = r2_score(df["成績"], model1.predict(df[["age"]]))
r2_v2 = r2_score(df_chaos["成績"], model2.predict(df_chaos[["age"]]))

print(f"原始模型 (v1) 的 R2 分數: {r2_v1:.4f}")
print(f"搗蛋後模型 (v2) 的 R2 分數: {r2_v2:.4f}")
```

```python
df_chaos['target'] = df_chaos['status'].apply(lambda x: 1 if x == "Pass" else 0)

print(df_chaos[['姓名', 'status', 'target']])
```

```python
clf = LogisticRegression()
clf.fit(df_chaos[["age"]] , df_chaos['target'])
print("邏輯回歸模型訓練完成")
```

```python
test_age = pd.DataFrame({"age":[50]})
pred = clf.predict(test_age)
prob = clf.predict_proba(test_age)

print(f"50 歲的預測結果：{'Pass' if pred[0] == 1 else 'Fail'}")
print(f"AI 認為 Pass 的機率是：{prob[0][1]*100:.2f}%")
```
