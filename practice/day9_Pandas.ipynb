```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "name": ["Eric", "gpt", "A", "B", "C"],
    "age": ["24", "10", "65", "66", "67"],
    "score": [420, 100, 0.56, 6.6, 76],
    "city": ["1", "2", "3", "4", "5"]
})
df
```

```python
df2 = pd.DataFrame({
    "name": ["Eric", "gpt", "A", "B", "C"],
    "level": ["S", "A", "A", "B", "C"]
})
df2
```

```python
merged = pd.merge(df, df2, on="name")
merged
```

```python
merged.groupby("level")["score"].mean()
```

```python
merged.groupby("level")["score"].agg(["mean", "max", "min"])
```
