```python
import numpy as np

arr = np.arange(1, 10).reshape(3, 3)
print("axis = 0", np.sum(arr, axis = 0))
print("axis = 1", np.sum(arr, axis = 1))
```

```python
print(np.max(arr, axis = 0))
print(np.max(arr, axis = 1))

print(np.mean(arr, axis = 0))
print(np.mean(arr, axis = 1))
```
