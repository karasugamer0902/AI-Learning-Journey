```python
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

model_big = joblib.load('100people_score_model.pkl')
df_big = pd.read_csv("data_big.csv")
```

```python
X_big = df_big[['age' , 'hours']]
y_big = df_big['score']

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_big, y_big, test_size=0.2, random_state=42)
```

```python
y_pred_b = model_big.predict(X_test_b)

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(8, 8))
plt.scatter(y_test_b, y_pred_b, color='green', alpha=0.6)

plt.plot([y_big.min(), y_big.max()], [y_big.min(), y_big.max()], 'r--', lw=2)

plt.title('預測值 vs 實際值 (理想應落在紅線上)')
plt.xlabel('實際成績')
plt.ylabel('AI 預測成績')
plt.grid(True)
plt.show()
```
