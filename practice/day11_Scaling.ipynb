```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("cleaned_data.csv")
raw
```

```python
sns.set_theme(style="whitegrid")
plt.Figure(figsize=(8, 4))

sns.histplot(raw["age"] , kde = True, color="orange")
plt.title("Age Distribution Analysis")
plt.show()
```

```python
raw["age_scaled"] = (raw["age"] - raw["age"].min()) / (raw["age"].max() - raw["age"].min())
print(raw[["age" , "age_scaled"]])
```
