```python
import pandas as pd
import joblib
from sklearn.linear_model import LinearRegression

model1 = joblib.load('linear_regression_v1.pkl')
model2 = joblib.load('linear_regression_v2.pkl')
df = pd.read_csv("cleaned_data.csv")
df_chaos = pd.read_csv("data_chaos.csv")
```

```python
df_chaos['hours'] = [10, 5, 2, 1, 4, 20]
df_chaos
```

```python
X_multi = df_chaos[["age" , "hours"]]
y_score = df_chaos["成績"]

model_multi = LinearRegression()
model_multi.fit(X_multi , y_score)

print(f"多特徵模型的 R2 分數: {model_multi.score(X_multi, y_score):.4f}")
```

```python
print(f"年齡的權重 (Age Coef): {model_multi.coef_[0]:.2f}")
print(f"讀書的權重 (Hours Coef): {model_multi.coef_[1]:.2f}")
```

```python
test_lazy = pd.DataFrame({"age":[20] , "hours":[0.5]})
lazy_pred = model_multi.predict(test_lazy)

print(f"20 歲但只讀 0.5 小時的預測分數：{lazy_pred[0]:.2f} 分")
```

```python
test_diligent = pd.DataFrame({"age": [80], "hours": [15]})
diligent_pred = model_multi.predict(test_diligent)

print(f"80 歲且讀書 15 小時的預測分數：{diligent_pred[0]:.2f} 分")
```
